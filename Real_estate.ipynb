```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep

PATH = 'D:\Learnings\web_driver\chromedriver.exe'

driver = webdriver.Chrome(executable_path=PATH)
driver.maximize_window()

driver.get('https://www.imoti.net/en/obiavi/r/prodava/evbeja/kashta/?sid=iAY4nY')

size = driver.find_elements(By.XPATH,"//*[@class='inline-group']/h3/span[2]")
rate = driver.find_elements(By.XPATH,"//ul/li[2]/strong")
price = driver.find_element(By.XPATH,"//div/header/strong")

prices = []
sizes = []
rates = []
nxt = driver.find_element(By.XPATH,"//a[@title='Next']")
rng = len(driver.find_elements(By.XPATH,'//nav[@class="paginator"]/a'))
for i in range(1,rng):
    print("Scraping page", i)
    price = driver.find_elements(By.XPATH,"//div/header/strong")
    size = driver.find_elements(By.XPATH,"//*[@class='inline-group']/h3/span[2]")
    rate = driver.find_elements(By.XPATH,"//ul/li[2]/strong")

    for j in price:
        prices.append(j.text)
    for k in size:
        sizes.append(k.text)
    for l in rate:
        rates.append(l.text)
    nxt.click()
    sleep(5)
```

```python
data = pd.DataFrame()
data['Price'] = prices
data['House Size'] = sizes
data['Area Rate'] = rates

data
```
